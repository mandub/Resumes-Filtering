```python
path2 = "Open source global resumes"
import sys,os
resumes=[]  # resumes[[ID][Source][Path][file_name][Extension][Resume_Object]]
Directories =[]
path= './' + path2
```

```python
# resumes[[ID][Source][Path][file_name][Extension][Resume_Object]]
def fileExtension(file_name):
    if 'pdf' in file_name:
        return 'pdf'
    elif 'docx' in file_name:
        return 'docx'
    elif 'doc' in file_name:
        return 'doc'
    else:
        return 'unknown'
ID = 0
for root, subFolders, files in os.walk(path ):#+'/Random'
    #print  (root)
    #print ('-------------------')
    #print files
    for file_name in files:
        #print ('X = ' + file_name)
        ID = ID + 1
        Source = 'Open source global resumes'
        resume = [ID,Source,root,file_name,fileExtension(file_name)]
        resumes.append(resume)
```

```python
import docx
def docx_to_text(resume):
    print (resume[2]+'/'+resume[3] )
    doc = docx.Document(resume[2]+'/'+resume[3])
    all_paras = doc.paragraphs
    text = []
    for para in all_paras:
        #text.append[str (para.text)]
        text.append(str (para.text))
        #print (str (para.text))
    return text

for i in resumes:
    if i[4] == 'docx':
        text = docx_to_text(i)
        i.append(text)
        print (i[0])


'''
print (resumes[0][2]+'/'+resumes[0][3] )
doc = docx.Document(resumes[0][2]+'/'+resumes[0][3])
all_paras = doc.paragraphs
text = []
print (len(all_paras))
for para in all_paras:
    #text.append[str (para.text)]
    text.append(str (para.text))
    print (str (para.text))
print (text)'''
```

```python

import PyPDF2
sample= resumes[1][2]+'/'+resumes[1][3]
pdf_file = open(sample, 'rb')
read_pdf = PyPDF2.PdfFileReader(pdf_file)
number_of_pages = read_pdf.getNumPages()
page = read_pdf.getPage(0)
page_content = page.extractText()
#print (str (page_content))# .encode('utf-8')
resumeString=''
for i in page_content:
    resumeString= resumeString + i

print (resumeString)
```

```python
import nltk
import numpy
from nltk.corpus import stopwords
from nltk.tokenize import sent_tokenize,word_tokenize
stop_words = set (stopwords.words("english"))
#print (stop_words)
#print (word_tokenize(resumeString))
words = word_tokenize(resumeString)
filter_resume = [w for w in words if not w in stop_words]
filter_resume
```
